## CONVERT SPECIES STRING INTO NUMBER USING MAPPING METHOD

```python
species_mapping = {
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
}

df1['Species'] = df1['Species'].replace(species_mapping)
```

```python
df1.head()
```

```python
df1.Species.value_counts()
```

```python
# SEPERATE X AND Y

x = df1.drop('Species', axis=1)
y = df1['Species']  
```

## PREPROCESSING 

```python
# SCALLING

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_sc=sc.fit_transform(x)
```

```python
# CROSS VALIDATION

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sc,y,test_size=0.2,random_state=11)
```

## LOGISTIC REGRESSION

```python
# MODEL TRAINING

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr_predi = lr.predict(x_test)
lr_predi
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print('Accuracy Score : ', accuracy_score(y_test, lr_predi))
```

```python
print('Confusion Matrix :', confusion_matrix(y_test, lr_predi))
```

```python
print('Classification Report :', classification_report(y_test,lr_predi))
```

USING THE METHOD OF LOGISTIC REGRESSION, WE CAN OBSERVE THAT THE ACCURACY OF IRIS FLOWER DATASET IS APPROXIMATELY 93.33%
