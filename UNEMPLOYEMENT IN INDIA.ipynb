```python
plt.figure(figsize = (8,4))
sns.histplot(df[' Estimated Unemployment Rate (%)'], bins = 30, color = 'purple')
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.histplot(df[' Estimated Employed'], bins = 40, color = 'red')
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.histplot(df[' Estimated Labour Participation Rate (%)'], bins = 25, color = 'salmon')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data = df, x = 'Region', y = ' Estimated Unemployment Rate (%)')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

FROM THE BAR-GRAPH, WE CAN OBSERVE THAT TRIPURA HAS THE HIGHEST UNEMPLOYEEMENT RATE.

```python
plt.figure(figsize=(10,4))
sns.barplot(data = df, x = 'Region', y = ' Estimated Unemployment Rate (%)', hue = 'Area')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data = df, x = 'Region', y = ' Estimated Employed')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

FROM THE BAR-GRAPH, WE CAN OBSERVE THAT UTTAR PRADESH HAS THE HIGJEST EMPLOYEEMENT RATE.

```python
plt.figure(figsize=(10,4))
sns.barplot(data = df, x = 'Region', y = ' Estimated Employed', hue = 'Area')
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```
